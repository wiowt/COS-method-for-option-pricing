```python
import numpy as np
import matplotlib.pyplot as plt

import COS as cos
```

```python
K = np.linspace(5, 300, 100)
S = 100
r = 0.05
T = 5
```

```python
model_gbm = cos.GBMModel(sigma=0.1)
```

```python
rnd = model_gbm.rnd(S, K, r, T)
plt.plot(K, rnd)
plt.grid()
plt.title("Risk-neutral density of GBM")
plt.show()
```

```python
model_heston = cos.HestonModel(kappa=1.5768, theta=0.0398, dzeta=0.05751, rho=-0.5711, v0=0.0175)

rnd = model_heston.rnd(S, K, r, T)
plt.plot(K, rnd)
plt.grid()
plt.title("Risk-neutral density of Heston model")
plt.show()
```

```python
model_cgmy = cos.CGMYModel(C=1, G=5, M=5, Y=0.5)

rnd = model_cgmy.rnd(S, K, r, T)
plt.plot(K, rnd)


plt.grid()
plt.title("Risk-neutral density of CGMY model")
plt.show()
```
